```python
# ============================================================
#  WASTE SEGREGATION AI AGENT (SIMPLE VERSION)
# ============================================================

import json
from typing import Dict

# ------------------------------------------------------------
# TOOL 1: Basic Waste Classifier (simple rule-based)
# ------------------------------------------------------------
def classify_waste(item: str) -> str:
    item = item.lower()

    recyclable = ["plastic", "bottle", "paper", "cardboard", "metal", "can"]
    organic = ["food", "banana", "vegetable", "fruit", "leftover"]
    hazardous = ["battery", "chemical", "paint", "e-waste"]

    if any(word in item for word in recyclable):
        return "Recyclable Waste"
    if any(word in item for word in organic):
        return "Organic / Biodegradable Waste"
    if any(word in item for word in hazardous):
        return "Hazardous Waste"

    return "General Waste"


# ------------------------------------------------------------
# TOOL 2: Disposal Instructions
# ------------------------------------------------------------
def disposal_instructions(category: str) -> str:
    instructions = {
        "Recyclable Waste": "Place in the dry waste bin. Keep it clean and dry.",
        "Organic / Biodegradable Waste": "Place in the green/organic bin.",
        "Hazardous Waste": "Place in the red bin. Handle carefully.",
        "General Waste": "Place in the general waste bin."
    }
    return instructions.get(category, "No instructions available.")


# ------------------------------------------------------------
# SIMPLE AI AGENT
# ------------------------------------------------------------
def waste_agent(user_input: str) -> Dict:
    """
    The agent decides which tool to call and produces a final answer.
    """

    # Step 1: Classify
    category = classify_waste(user_input)

    # Step 2: Get disposal steps
    steps = disposal_instructions(category)

    # Step 3: Structure the agent response
    return {
        "item": user_input,
        "category": category,
        "instructions": steps
    }


# ------------------------------------------------------------
# TEST (You can change the item)
# ------------------------------------------------------------
query = "banana peel"
result = waste_agent(query)
result
```
